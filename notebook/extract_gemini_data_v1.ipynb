```python
# search urls for each pages
all_urls = links
for url in links:
    all_urls = all_urls + get_page_urls(url, url_keywords)
    time.sleep(1)
print('find all urls at depth 1')
```

```python
# parse links
unique_urls = list(set(all_urls))
print(unique_urls)
```

```python
# depth2 urls
depth2_urls = unique_urls
for url in unique_urls:
    try:
        depth2_urls = depth2_urls + get_page_urls(url, url_keywords)
    except:
        print(url)
    time.sleep(1)
print('find all urls at depth 2')
```

```python
# remove certain type of urls
remove_keywords = ['linkedin', 'twitter', 'facebook', 'youtube']
depth2_urls_unique_clean = []
for url in depth2_urls_unique:
    status = True
    for value in remove_keywords:
        if value.lower() in url.lower():
            status = False
            
    if status:
        depth2_urls_unique_clean.append(url)
```

```python
def download_html(url, index):
    response = requests.get(url)
    html     = response.text
    os.makedirs(os.path.join(output_dir, 'html_v2'), exist_ok=True)
    path = os.path.join(output_dir, 'html_v2', f"{index}_{url.split('/')[-1]}.html")
    path = path.replace("=", '').replace("?", '')
    print(path)
    if os.path.exists(path):
        return path
    with open(path, 'w', encoding='utf-8') as f:
        f.write(html)
    time.sleep(0.5)
    return path
```

```python
paths = []
for index, value in enumerate(depth2_urls_unique_clean):
    paths.append(download_html(value, index))
```

```python
df_res = pd.DataFrame()
df_res['link']    = depth2_urls_unique_clean
df_res['path']    = paths
```

```python
df_res
```

```python
def clean_html(path):
    with open(path, 'r', encoding='utf-8') as f:
        html = f.read()
        
#     print(html)
    soup = BeautifulSoup(html, 'html.parser')    
    try:
        text = soup.get_text()
    except:
        text = ''
    
    text = text.strip()
    text = text.replace('Gemini Solutions | IT Consulting, Product DevelopmentAbout UsSolutionsTechnologiesGemUCareersContact', '')
    text = text.replace('Data ScienceDevSecOpsQuality EngineeringTechnical InfrastructureData EngineeringCloud SolutionsApplication DevelopmentDigital OfferingsManaged ServicesCloud Solutions', '')
    text = text.replace('Data ScienceDevSecOpsQuality EngineeringTechnical InfrastructureData EngineeringCloud SolutionsApplication DevelopmentDigital OfferingsManaged Services', '')
    text = text.replace('LinksCopyright © 2023 Gemini Solutions Pvt. ltd. All rights reserved.Terms & Conditions|Privacy Policy', '')
    text = text.replace('Solutions Application Development Digital Offerings Managed Services Cloud Solutions Data Engineering Data Science  Technical Infrastructure DevSecOps Quality EngineeringTechnologies Front-End  Core Development Databases Cloud BI & Analytics MobilityPeople About Us Meet Our Team [email\xa0protected] Careers Campus Connect Hiring ProcessQuick Links Blogs Case Studies Contact UsSolutionsTechnologiesPeopleQuick', '')
    text = text.replace('Digital Offerings Managed Services Cloud Solutions Data Engineering Data Science  Technical Infrastructure DevSecOps Quality EngineeringTechnologies Front-End  Core Development Databases Cloud BI & Analytics MobilityPeople About Us Meet Our Team [email\xa0protected] Careers Campus Connect Hiring ProcessQuick Links Blogs Case Studies Contact UsSolutionsTechnologiesPeopleQuick', '')
    text = re.sub(r'(\n\s*)+\n+', '\n\n', text)
    text = text.strip()
    return text
```

```python
df_res['content'] = df_res['path'].apply(lambda x: clean_html(x))
print('text extraction complete')
```
